# 1.Fetching data

```python
from urllib.request import urlopen 
```

```python
android_url="https://en.wikipedia.org/wiki/Android_version_history"
```

```python
android_data=urlopen(android_url)
print(type(android_data))
```

```python
android_html=android_data.read()

android_data.close()
```

# 2.Parsing data

```python
from bs4 import BeautifulSoup as soup
```

```python
android_soup=soup(android_html,"html.parser")
print(android_soup)
```

```python
print(type(android_soup))
```

```python
tables=android_soup.findAll("table",{"class":"wikitable"})
tables
```

```python
print(len(tables))
```

```python
android_tables=tables[0]
android_tables
```

# 3.Parsing HTML
### .Remove undesired tags
### .Extract table header and tag¶

```python
headers=android_tables.findAll("th")
print(len(headers))
```

```python
print(headers)
```

```python
print(headers[0].text)
```

```python
column_titles=[ct.text[:-1] for ct in headers]
print(column_titles)
```

```python
row_data=android_tables.findAll("tr")[1:]
print(len(row_data))
first_row=row_data[0].findAll("td")
for d in first_row:
    print(d.text[:-1])
```

```python
table_rows=[]
for row in row_data:
    current_row=[]
    row_data=row.findAll("td",{})
    for idx,data in enumerate(row_data):
            current_row.append(data.text[:-1])
    table_rows.append(current_row)
print(table_rows)
```

# 4.Creating a CSV file

```python
file_name="android_history_version.csv"
with open(file_name,"w",encoding="utf-8") as f:
    
    #Write the header
    header_string=",".join(column_titles)  #Creating a string bcoz in CSV each line is a string
    header_string +="\n"
    f.write(header_string)
    
    for row in table_rows[:-1]:
        row_string=""
        for w in row:
            w=w.replace(",","")
            row_string+=w+","
        row_string=row_string[:-1]
        row_string+="\n"
        f.write(row_string)
```
