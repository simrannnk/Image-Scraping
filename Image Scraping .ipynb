```python
import requests
import bs4
```

```python
url="https://www.passiton.com/inspirational-quotes"
```

```python
response=requests.get(url)
response.content
```

```python
#HTML Parser
soup=bs4.BeautifulSoup(response.content)   #Parsed HTML response to a python object format 
```

```python
image=soup.findAll("img")
image[9]
```

```python
image[9].attrs
```

```python
image[9].attrs["src"]
```

```python
with open("inspirational.jpg","wb") as file:
    img_url=image[9].attrs["src"]
    response=requests.get(img_url)
    
    file.write(response.content)
```
